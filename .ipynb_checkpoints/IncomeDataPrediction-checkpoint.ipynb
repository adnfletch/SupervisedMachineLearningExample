The goal of this project is to evaluate labeled data to predict if an individual's annual income is over 50k based off of demographic data that is both numerical and categorical. Used a simple Logistic Regression model

```python
import sklearn
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Import the data and examine the categories:
data = pd.read_csv("income_cleaned.csv")
data.info()
data.head()
```

The income column is a binary 0 or 1 indicating above or below 50k. There are 29750 entries. The next block cleans any null or incorrect data from the dataframe. We will find that no data is null, we will keep all of the data.

```python
data.isna().sum(0)
```

```python
#For interest, check the distributions of numerical data and their relationships as an initial investigation:
import seaborn as sns
sns.pairplot(data[["education-num", "hours-per-week", "age"]])
plt.show()
#Will notice that there are no clear relationships between the quantitative features.
```

```python
#Create a count plot:
sns.countplot(x="income_50k", data = data)
plt.show()
```

```python
#split the categorical variables into multiple true or false columns for each option. Rather than encoding (as there is not a true natural order between the categories)
encoded_cols = pd.get_dummies(data=data[['workclass', 'occupation', 'race', 'sex']], prefix = 'encoded_')
encoded_cols.info()
```

```python
#replace the categorical columns in the original dataframe with the new binary columns. 
data = data.drop(labels = ["workclass","occupation", "race", "sex"], axis=1)
data = pd.concat([encoded_cols, data], axis = 1) #Axis = 1 means it is concating along columns rather than between rows.
data.info()
```

```python
#Now split the data into a test set and a training set
from sklearn.model_selection import train_test_split

y = data["income_50k"] #The predicted variable
x = data.drop(["income_50k"], axis=1) #all of the feature columns

#randomly split into a train and test set. Arbitrarily selected 20% to be used for training
#ramdon state is used so that every time the dataset is split, reproducable results are found.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1) 
```

```python
#train a logistic regression model using the basic features
from sklearn.linear_model import LogisticRegression
multi_feature_log_model = LogisticRegression()
multi_feature_log_model.fit(x_train, y_train)
```

```python
#Notice that the solver did not converge. try again using a different solver
multi_feature_log_model = LogisticRegression(solver = "newton-cg")
multi_feature_log_model.fit(x_train, y_train)
```

```python
#try predicting, and check the models results:
log_predictions = multi_feature_log_model.predict(x_test)
```

```python
#Visualize predictions
from sklearn import metrics

matrix = metrics.confusion_matrix(y_test, log_predictions)
disp = metrics.ConfusionMatrixDisplay(matrix)
disp.plot()
```

This is a confusion matrix. Shows how many true positives, true negatives, false positives, false negatives. So there were 1529 true positives, 743 false negatives. 3159 True negatives, 519 false positives. eg: If this were a medical application, we would want a low number of false negatives, you would rather have something flagged for a doctor than missing it. 

```python
accuracy = metrics.accuracy_score(y_true = y_test, y_pred = log_predictions)
print("Overall prediction accuracy as a decimal: ", accuracy)
```
